```python
import pandas as pd
import json
```

```python
df = pd.read_csv('tcp_zip.csv')
```

```python
dic = {}

for i in range(len(df)):
    dic[int(df.iloc[i,0])] = int(df.iloc[i,0])
```

```python
zip_json = json.dumps(dic,sort_keys=False,separators=(',', ': '))
```

```python
print(type(zip_json))
```

```python
f = open('zip_json.json', 'w')
```

```python
f.write(zip_json)
```

```python
ca_zip = json.load(open('zip_json.json'))
```
